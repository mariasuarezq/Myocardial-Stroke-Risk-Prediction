## EXPLORATORY DATA ANALYSIS (EDA)

Exploratory data analysis (EDA) is a very important step because it is very important to be able to understand the nature of the data without making assumptions. The results of data exploration can be extremely useful in grasping the structure of the data, the distribution of the values, and the presence of extreme values and interrelationships within the data set. The purpose of EDA is to use summary statistics and visualizations to better understand data, find clues about the tendencies of the data.

Let's start again by loading the libraries and by loading the DataSet into a DataFrame:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
stroke = pd.read_csv(r"C:\Users\maria\Desktop\proyecto infarto de miocardio\healthcare-dataset-stroke-data.csv")
stroke = stroke.dropna(subset=['bmi'])
stroke.head()
```

**Data visualization using Boxplots and Bar Charts**

We are going to create boxplots and bar charts to see if there is a relationship between different features and the probability of stroke:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='stroke', y='id', data=stroke)
plt.title('Boxplot of ID in relation to Stroke')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='gender', hue='stroke', data=stroke)
plt.title('Bar chart of gender in relation to Stroke')
plt.show()
```

Let's look at the percentages to have a clearer visualization:

```python
plt.figure(figsize=(10, 6))

total_por_categoria = stroke['gender'].value_counts()
porcentaje_con_infarto = (stroke[stroke['stroke'] == 1]['gender'].value_counts() / total_por_categoria) * 100

porcentajes_stroke = pd.DataFrame({'gender': porcentaje_con_infarto.index, 'porcentaje_con_infarto': porcentaje_con_infarto.values})

porcentajes_stroke = porcentajes_stroke.sort_values(by='gender')

sns.barplot(x='gender', y='porcentaje_con_infarto', data=porcentajes_stroke, hue='gender')

plt.title('Percentage of people with stroke by gender categor')
plt.ylabel('Percentage with stroke')
plt.xlabel('gender')
plt.show()
```

**Heatmap**

Let's create a heatmap to visualize if there is correlation between different features (excluding non-continuous features):

```python
plt.figure(figsize=(20, 12)) 
columns_to_drop = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
sns.heatmap(stroke.drop(columns=columns_to_drop).corr(), annot=True)
plt.show()
```

It is clear that the diagonal has a correlation coefficient of 1 because it represents the correlation between a variable and itself. We observe a moderate correlation between BMI and age. Let's create a graph to represent the correlation between these variables:

```python
cols = ['id', 'age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi', 'stroke']
sns.pairplot(stroke[cols])
```

Again, we observe only a moderate correlation between age and BMI. Let's move on to data preprocessing.
